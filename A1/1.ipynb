**ETT dataset**
---



```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
ETTh1 = pd.read_csv("/content/all_six_datasets/ETT-small/ETTh1.csv")
ETTh2 = pd.read_csv("/content/all_six_datasets/ETT-small/ETTh2.csv")
ETTm1 = pd.read_csv("/content/all_six_datasets/ETT-small/ETTm1.csv")
ETTm2 = pd.read_csv("/content/all_six_datasets/ETT-small/ETTm2.csv")
```

```python
ETTh1.head()
```

**convert date column to date_time column and set it as the index**

```python
ETTh1['date'] = pd.to_datetime(ETTh1['date'])
ETTh1 = ETTh1.set_index('date')
ETTh2['date'] = pd.to_datetime(ETTh2['date'])
ETTh2 = ETTh2.set_index('date')
ETTm1['date'] = pd.to_datetime(ETTm1['date'])
ETTm1 = ETTm1.set_index('date')
ETTm2['date'] = pd.to_datetime(ETTm2['date'])
ETTm2 = ETTm2.set_index('date')
```

```python
def plot_ett_series(df, title):
    axes = df.plot(y='OT', figsize=(16,10), legend=True, sharex=False, title=title)
```

**Plot ETTh1 dataset**

```python
plot_ett_series(ETTh1, "ETTh1")
```

**Plot ETTh2 dataset**

```python
plot_ett_series(ETTh2, "ETTh2")
```

**Plot ETTm1 dataset**

```python
plot_ett_series(ETTm1, "ETTm1")
```

**Plot ETTm2 dataset**

```python
plot_ett_series(ETTm2, "ETTm2")
```
